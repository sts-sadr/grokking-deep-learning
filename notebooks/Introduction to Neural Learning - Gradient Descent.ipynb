# What's the Simplest Form of Neural Learning?

```python
weight = .1
lr = 0.01
def neural_network(input, weight):
    prediction = input * weight
    return prediction
```

```python
X = [8.5]
y = [1]

input = X[0]
true = y[0]

pred = neural_network(input, weight)
error = (pred - true) ** 2
error = round(error, 4)
print(error)
```

```python
lr = .1
p_up = neural_network(input, weight+lr)
e_up = (p_up - true) ** 2
e_up = round(e_up, 4)
print(e_up)
```

```python
lr = .01
p_dn = neural_network(input, weight-lr)
e_dn = (p_dn - true) ** 2
e_dn = round(e_dn, 4)
print(e_dn)
```

# Hot and Cold Learning 

```python
weight = .5
input = .5
goal_prediction = .8

step_amount = 0.001

for iteration in range(1101):
    prediction = input * weight
    error = (prediction - goal_prediction) ** 2
    
    print("Error: {} Prediction: {}".format(round(error,4), round(prediction, 4)))
    
    up_prediction = input * (weight + step_amount)
    up_error = (up_prediction - goal_prediction) ** 2
    
    down_prediction = input * (weight - step_amount)
    down_error = (down_prediction - goal_prediction) ** 2
    
    if down_error < up_error:
        weight = weight - step_amount
    if up_error < down_error:
        weight = weight + step_amount
```

# Calculating Both direction and amount from error

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = input * weight
    error = (goal_pred - pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount
    print("Error: {} Prediction: {}".format(round(error,4), round(pred, 4)))
```
